 # Wikipedia Vote Network

```python
!wget https://snap.stanford.edu/data/wiki-Vote.txt.gz
!gunzip wiki-Vote.txt.gz
```

```python
wiki_vote_df = pd.DataFrame(np.loadtxt('wiki-Vote.txt', dtype='int'))
```

```python
wiki_vote_df
```

```python
G = nx.convert_matrix.from_pandas_edgelist(wiki_vote_df, 0, 1)
```

```python
pos = nx.layout.spring_layout(G)
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
nx.draw(G, pos=pos, node_size=10, ax=ax)
ax.set_title('Wikipedia Voter Network')
fig.show()
```

This graph has a dense core of Wikipedia users that voted for one another, and a slight outer ring of a few users that voted for one another. No one from the dense core voted for someone in the outer ring and vice versa.
